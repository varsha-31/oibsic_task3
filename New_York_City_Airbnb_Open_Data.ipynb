```python
import pandas as pd
# Load the dataset
df = pd.read_csv('/content/AB_NYC_2019.csv')
df
```

```python
# Check for missing values
print(df.isnull().sum())

# Summary statistics
print(df.describe())
```

```python
# Drop rows with missing values
df.dropna(inplace=True)

# Impute missing values
# Example: Replace missing values in a column with the mean
df['latitude'].fillna(df['latitude'].mean(), inplace=True)
```

```python
# Check for duplicates
print(df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert data types
df['latitude'] = pd.to_numeric(df['latitude'])
df['price'] = pd.to_datetime(df['price'])
```

```python
# Example: Check if a column has values within a specific range
valid_range = (0, 100)
invalid_values = df[~df['latitude'].between(*valid_range)]
```

```python
# Example: Check if a column has any negative values (if it shouldn't)
negative_values = df[df['latitude'] < 0]
negative_values
```
